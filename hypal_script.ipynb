```python
import pandas as pd
import numpy as np
import textwrap
import re
```

```python
excel_path = input('Enter the path of the file "hypal_search_export_w-metadata.xlsx": ')
df = pd.read_excel(excel_path, index_col=0)
```

```python
# Converting document_id from index to column

df['document_id'] = df.index
df = df.reset_index(drop=True)

# Moving document_id to the beginning of the dataframe

first_column = df.pop('document_id')
df.insert(0, 'document_id', first_column)

# Storing the TT and the ST in a separate dataframe

df2 = pd.DataFrame()
df2['ST segment'] = df['ST segment']
df2['TT segment'] = df['TT segment']

# Keeping only metadata in the first dataframe

df = df.drop('ST segment (formatted)', axis=1)
df = df.drop('TT segment (formatted)', axis=1)
df = df.drop('ST segment', axis=1)
df = df.drop('TT segment', axis=1)
```

```python
selected_tags = []

def select_tags(selected_tags):

  options = []
  options_without_formatting = []

  col_option = 0

  for col in df.columns.values.tolist():
    col_string = '[' + str(col_option) + ']' + ' ' + col
    col_option += 1
    options.append(col_string)
    options_without_formatting.append(col)

  options_str = ' '.join(options)

  print ('\n')

  print(textwrap.fill(options_str, 160))

  print('\n')

  x = list(map(int, input('Select metatags (only numbers, separated by whitespace): ').split()))

  opts = []
  opts_without_formatting = []

  for num in x:
    opts.append(options[num])
    selected_tags.append(options_without_formatting[num])
  
  ops_str = ' '.join(opts)

  print('\n')

  print('Selected metatags: ' + ops_str)
```

```python
select_tags(selected_tags)
```

```python
all_tags = df.columns.values.tolist()
tags_to_remove = [x for x in all_tags if x not in selected_tags]
df = df.drop(columns=tags_to_remove)
```

```python
df
```

```python
# create an empty list to hold the xml strings
xml_strings = []

# iterate through each row of df2
for i, row in df2.iterrows():
    # get the value in the TT segment column
    tt_segment = row['TT segment']
    
    # create the xml string
    xml_string = '<doc '
    
    # iterate through each column of df
    for col in df.columns:
        # get the value in the corresponding row and column of df
        val = str(df.loc[i, col])
        
        # append the column name and value to the xml string
        xml_string += col + '="' + val + '" '
    
    # close the xml tag and append the TT segment text
    xml_string += '>' + tt_segment + '</doc>'
    
    # append the xml string to the list
    xml_strings.append(xml_string)

# replace the content of the TT segment column with the xml column
df2['TT segment'] = xml_strings
```

```python
df2
```

```python
# Regex

regex1 = '\[\/mark\] \[mark\]'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex1, '', x))

regex2 = '\[\/mark\]'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex2, '</mark>', x))

regex3 = '\[mark\]'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex3, '<mark>', x))

regex4 = '\] <'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex4, '> <', x))

regex5 = '\]'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex5, '>', x))

regex6 = '\[\/'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex6, '</', x))

regex7 = '\['
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex7, '<', x))

regex8 = '\-'
df2['TT segment'] = df2['TT segment'].apply(lambda x: re.sub(regex8, '_', x))
```

```python
df2 = df2.rename(columns={"ST segment": "English", "TT segment": "Italian"})
```
